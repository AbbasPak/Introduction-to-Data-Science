## Matplotlib Object Oriented Method

Using Matplotlibâ€™s **Object-Oriented (OO) API** provides more control over plot elements, making it ideal for complex or customized visualizations. 


### Key Elements of the Object-Oriented API
1. **Figure**: The overall container for the plots.
2. **Axes**: The actual plotting area where data is plotted (think of it as a subplot).
3. **Axis**: The x and y axis within each Axes object.

### Basic Steps with Examples

#### Step 1: Create a Figure and Axes
You create a figure (the overall window) and one or more axes (subplots) within it.

```python


import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Figure and an Axes object
fig, ax = plt.subplots()  # This creates one Axes object
ax.plot(x, y)

# Customize
ax.set_title("Sine Wave")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```

#### Step 2: Multiple Plots on the Same Canvas

To have multiple subplots, you can specify the number of rows and columns in `subplots()`.

```python



fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of plots

# Plot on each Axes individually
axs[0, 0].plot(x, np.sin(x), 'r')
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, np.cos(x), 'b')
axs[0, 1].set_title("Cosine")

axs[1, 0].plot(x, np.tan(x), 'g')
axs[1, 0].set_title("Tangent")

axs[1, 1].plot(x, -np.sin(x), 'm')
axs[1, 1].set_title("Negative Sine")

# Adjust layout
fig.tight_layout()
plt.show()
```


#### Step 3: Customizing Each Plot Further
The OO approach allows you to access and customize each part of the plot. Hereâ€™s an example of modifying tick parameters and adding gridlines.


```python

fig, ax = plt.subplots(figsize=(8, 5))

# Plot with customizations
ax.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o', markersize=5)
ax.set_title("Customized Sine Wave")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Customize ticks and grid
ax.tick_params(axis='both', which='major', labelsize=12, color='blue', direction='inout')
ax.grid(True, linestyle='-', linewidth=0.5)

plt.show()
```

```python
#### Step 4: Using `GridSpec` for More Complex Layouts
#If you need irregular layouts, `GridSpec` gives finer control over subplot arrangement.


from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 8))
gs = GridSpec(3, 3, figure=fig)

# Define the layout
ax1 = fig.add_subplot(gs[0, :2])  # First row, span two columns
ax2 = fig.add_subplot(gs[1:, 0])  # Left side, span two rows
ax3 = fig.add_subplot(gs[1, 1:])  # Middle, top half
ax4 = fig.add_subplot(gs[2, 1:])  # Middle, bottom half

# Add data to each Axes
ax1.plot(x, np.sin(x), 'r')
ax2.plot(x, np.cos(x), 'b')
ax3.plot(x, np.tan(x), 'g')
ax4.plot(x, -np.sin(x), 'm')

# Titles for each plot
ax1.set_title("Sine Wave")
ax2.set_title("Cosine Wave")
ax3.set_title("Tangent Wave")
ax4.set_title("Negative Sine Wave")

plt.tight_layout()
plt.show()
```






The Object-Oriented API allows for precise control, making it powerful for detailed and professional visualizations, especially with more complex layouts.
